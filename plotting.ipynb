```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
nmi_10x = pd.read_table('nmi_10x_pooled_8k_uncurl_net.tsv')
nmi_10x_2 = pd.read_table('nmi_10x_pooled_8k_uncurl_net_2.tsv')
```

```python
nmi_10x_2
```

```python
nmi_tasic = pd.read_table('nmi_tasic_uncurl_net.tsv')
nmi_tasic_2 = pd.read_table('nmi_tasic_uncurl_net_2.tsv')
```

```python
nmi_tasic_2
```

```python
nmi_zeisel = pd.read_table('nmi_zeisel_uncurl_net.tsv')
nmi_zeisel_2 = pd.read_table('nmi_zeisel_uncurl_net_2.tsv')
```

```python
nmi_zeisel_2
```

This is essentially "de-normalizing" the table.

```python
new_table = {'method': [], 'dataset': [], 'value': []}
```

```python
for i, row in nmi_10x_2.iterrows():
    for k, val in zip(row.keys(), row):
        new_table['method'].append(k)
        new_table['dataset'].append('10x_8k')
        new_table['value'].append(val)
for i, row in nmi_10x.iterrows():
    for k, val in zip(row.keys(), row):
        new_table['method'].append(k)
        new_table['dataset'].append('10x_8k')
        new_table['value'].append(val)
```

```python
for i, row in nmi_tasic_2.iterrows():
    for k, val in zip(row.keys(), row):
        new_table['method'].append(k)
        new_table['dataset'].append('tasic')
        new_table['value'].append(val)
for i, row in nmi_tasic.iterrows():
    for k, val in zip(row.keys(), row):
        new_table['method'].append(k)
        new_table['dataset'].append('tasic')
        new_table['value'].append(val)
```

```python
for i, row in nmi_zeisel_2.iterrows():
    for k, val in zip(row.keys(), row):
        new_table['method'].append(k)
        new_table['dataset'].append('zeisel')
        new_table['value'].append(val)
for i, row in nmi_zeisel.iterrows():
    for k, val in zip(row.keys(), row):
        new_table['method'].append(k)
        new_table['dataset'].append('zeisel')
        new_table['value'].append(val)
```

```python
new_table = pd.DataFrame(new_table)
```

```python
new_table_subset = new_table[(new_table.method == 'LogNorm_UncurlNetW_argmax')\
                             | (new_table.method == 'Poisson_W_argmax')]
```

```python
new_table_subset
```

```python
sns.set()
sns.set(font_scale=1.5)
```

```python
plt.figure(figsize=(16,9))
sns.barplot(x='dataset', y='value', hue='method', data=new_table_subset)
plt.ylim(0, 1)
plt.title('Argmax NMI of UncurlNet vs Uncurl for 3 datasets')
plt.savefig('nmi_plots_2.png')
```

```python
new_table_subset_2 = new_table[(new_table.method == 'LogNorm_UncurlNetW_tsne_km')\
                             | (new_table.method == 'Poisson_W_tsne_km')]
```

```python
plt.figure(figsize=(16,9))
sns.barplot(x='dataset', y='value', hue='method', data=new_table_subset_2)
plt.ylim(0, 1)
plt.title('TSNE + KM NMI of UncurlNet vs Uncurl for 3 datasets')
plt.savefig('nmi_plots_tsne_km.png')
```

```python
# plot different UncurlNet parameters
new_table_subset_3 = new_table[(new_table.method.str.contains('LogNorm_UncurlNet') &\
                                new_table.method.str.contains('argmax'))\
                             | (new_table.method == 'Poisson_W_argmax')]
```

```python
plt.figure(figsize=(16,9))
sns.barplot(x='dataset', y='value', hue='method', data=new_table_subset_3)
plt.ylim(0, 1)
plt.title('Argmax NMI of UncurlNet vs Uncurl for 3 datasets, with various parameter settings')
```
