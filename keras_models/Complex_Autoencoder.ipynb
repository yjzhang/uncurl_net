```python
from keras.layers import Input, Dense
from keras.models import Model
import keras.backend as K
import scipy.io
import numpy as np
import matplotlib.pyplot as plt
import uncurl
import umap
from tsne import bh_sne
```

```python
mat = scipy.io.loadmat('../data/10x_pooled_400.mat')
```

```python
dat = np.log10(1 + mat['data'].toarray().astype(np.float32))
genes = uncurl.max_variance_genes(dat, nbins=5, frac=0.2)
dat = dat[genes,:]
dat = dat.transpose()
(c,g) = dat.shape
print dat.shape
```

```python
enc_dim1 = 250 
enc_dim2 = 50 
enc_dim3 = 25


inp1 = Input(shape=(g,))
inp2 = Input(shape=(enc_dim1,))
inp3 = Input(shape=(enc_dim2,))


enc_1 = Dense(enc_dim1, activation='sigmoid')(inp1)
enc_2 = Dense(enc_dim2, activation='sigmoid')(enc_1)
enc_3 = Dense(enc_dim3, activation='sigmoid')(enc_2)
dec_2 = Dense(enc_dim2, activation='sigmoid')(enc_3)
dec_1 = Dense(enc_dim1, activation='sigmoid')(dec_2)
decoded = Dense(g, activation='sigmoid')(dec_1)

# this model maps an input to its reconstruction
autoencoder = Model(inp1, decoded)

# this model maps an input to its encoded representation
encoder1 = Model(inp1, enc_3)
```

```python
autoencoder.compile(optimizer='adadelta', loss='mean_squared_error')
```

```python
autoencoder.fit(dat, dat,
                epochs=50,
                batch_size=100,
                shuffle=True,
                validation_data=(dat, dat))
```

```python
encoded_imgs = encoder1.predict(dat)
```

```python
plt.scatter(encoded_imgs[:,0],encoded_imgs[:,1], c = mat['labels'][0])
```

```python
embedding = umap.UMAP(n_neighbors=5,
                      min_dist=0.3,
                      metric='correlation').fit_transform(encoded_imgs)
embedding.shape
```

```python
plt.scatter(embedding[:,0],embedding[:,1], c = mat['labels'][0], cmap = 'tab20b')
```

```python
X_2d = bh_sne(encoded_imgs.astype('float64'))
X_2d.shape
plt.scatter(X_2d[:,0],X_2d[:,1], c = mat['labels'][0], cmap = 'tab20b')
```
